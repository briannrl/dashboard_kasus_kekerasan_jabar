### Cek Jumlah Kasus Setiap Database
Tujuan:
- Melihat konsistensi jumlah kasus (groupby daerah dan tahun) untuk dataset: Usia, Pendidikan, Pekerjaan

```python
pd.options.display.max_rows = 999
```

```python
(usia.groupby(["kode_kabupaten_kota", "tahun"]).agg({"jumlah_korban":"sum"}).reset_index()
 .assign(jumlah_korban_pendidikan=lambda x: edu.groupby(["kode_kabupaten_kota", "tahun"]).agg({"jumlah":"sum"}).reset_index()["jumlah"],
        jumlah_korban_pekerjaan=lambda x: job.groupby(["kode_kabupaten_kota", "tahun"]).agg({"jumlah_korban":"sum"}).reset_index()["jumlah_korban"],
        usia_pdd=lambda x: x["jumlah_korban"] == x["jumlah_korban_pendidikan"],
        pdd_pkj=lambda x: x["jumlah_korban_pendidikan"] == x["jumlah_korban_pekerjaan"],
        nilai_sama=lambda x: x[["usia_pdd", "pdd_pkj"]].all(axis=1))
 .rename(columns={"jumlah_korban":"jumlah_korban_usia"})
 .drop(columns=["usia_pdd", "pdd_pkj"])
)
```

```python
(usia.groupby(["kode_kabupaten_kota", "tahun"]).agg({"jumlah_korban":"sum"}).reset_index()
 .assign(jumlah_korban_pendidikan=lambda x: edu.groupby(["kode_kabupaten_kota", "tahun"]).agg({"jumlah":"sum"}).reset_index()["jumlah"],
        jumlah_korban_pekerjaan=lambda x: job.groupby(["kode_kabupaten_kota", "tahun"]).agg({"jumlah_korban":"sum"}).reset_index()["jumlah_korban"],
        usia_pdd=lambda x: x["jumlah_korban"] == x["jumlah_korban_pendidikan"],
        pdd_pkj=lambda x: x["jumlah_korban_pendidikan"] == x["jumlah_korban_pekerjaan"],
        nilai_sama=lambda x: x[["usia_pdd", "pdd_pkj"]].all(axis=1))
 .rename(columns={"jumlah_korban":"jumlah_korban_usia"})
 .drop(columns=["usia_pdd", "pdd_pkj"])["nilai_sama"]
 .value_counts()
)
```

__Ada 15 row yang memiliki perbedaan jumlah korban berdasarkan tahun dan daerah dari dataset jumlah korban berdasarkan usia, pendidikan, dan pekerjaan.__

### Growth Jumlah Kekerasan dari Tahun Sebelumnya untuk Setiap Daerah

__Data di bawah bertujuan untuk mencari growth kasus kekerasan dibandingkan tahun sebelumnya untuk setiap daerah dan akan diekspor ke file csv untuk digunakan di Tableau.__

```python
(job
 .groupby(["tahun", "nama_kabupaten_kota"])
 .agg({"jumlah_korban":"sum"})
 .reset_index()
 .pivot_table(columns="tahun", index="nama_kabupaten_kota", values="jumlah_korban", aggfunc="sum")
 .reset_index()
 .assign(growth2019=lambda df_: (df_[2019]-df_[2018])/df_[2018]
        ,growth2020=lambda df_: (df_[2020]-df_[2019])/df_[2019]
        ,growth2021=lambda df_: (df_[2021]-df_[2020])/df_[2020])
 .melt(id_vars=["nama_kabupaten_kota"], var_name="growth_cat", value_vars=["growth2019", "growth2020", "growth2021"], value_name="growth_value")
 .assign(tahun=lambda x: x["growth_cat"].str.replace("growth", "").astype(int))
 .drop(columns="growth_cat")
)
```
